```python
import matplotlib
matplotlib.use('nbagg')
import matplotlib.animation as anm
import matplotlib.pyplot as plt
import math
import matplotlib.patches as patches
import numpy as np
import time
import random
import pandas as pd
import copy
from datetime import datetime

seed = 323
random.seed(seed)
```

```python
class RUATM:                                                                                            #UATM 0=A 1=B 2=C 3=D
    def __init__(self, airspace_max):
        self.airspace_max = airspace_max             #  10, 10, 10, 10
        
    
class DCB:
    def __init__(self, airspace_max): #DCBを行う関数
        self.airspace_max = airspace_max
        self.total_delay = np.zeros(4) #総遅延時間
        self.total_air_delay = np.zeros(4)
        self.num_delay = np.zeros((0, 4))
        
        
    def nonDCB(self, fpl_data, flight_data):   #DCBを行わずに空中遅延で容量を超えないように処理する（new_fplを出力）
        self.new_fpl_data = fpl_data.copy()
        self.flight_data = flight_data.copy()
        self.new_flight_data = flight_data.copy()
        
        for t in range(len(self.flight_data)):
            self.aircraft_num = np.array([np.count_nonzero(self.new_flight_data[t] == 0), 
                                          np.count_nonzero(self.new_flight_data[t] == 1), 
                                          np.count_nonzero(self.new_flight_data[t] == 2), 
                                          np.count_nonzero(self.new_flight_data[t] == 3)])
            
            while np.any((self.aircraft_num > self.airspace_max)): #遅延させたことでほかのruatmでキャパオーバーが発生するかもしれない。全部なくなるまでループ
                for u in range(4): #ruatmごとにcapaオーバーがあるか確認
                    if self.aircraft_num[u] > self.airspace_max[u]: #各UATMでcapacityを超えていないか確認。超えていたら保留
                        self.add_flight_id = np.where((self.new_flight_data[t] != self.new_flight_data[t-1]) 
                                                      & (self.new_flight_data[t] == u)) #RUATM:u にやってきたフライトの番号
                        self.add_flight_id = np.array(self.add_flight_id[0]) #あと処理

                        for index, i in enumerate(self.add_flight_id): #add_flightは複数かも
                            if index >= self.aircraft_num[u] - self.airspace_max[u]: break
                                
                            self.new_flight_data[t:-1, i] = self.new_flight_data[t-1:-2, i] #t以降のフライト計画を1秒ずつ遅らせる
                            
                            if self.new_fpl_data.iat[i, 1] == u: #depのときはdep, mid, app, arrを遅延指示
                                self.new_fpl_data.iat[i, 4] += 1
                                self.new_fpl_data.iat[i, 5] += 1
                                self.new_fpl_data.iat[i, 6] += 1
                                self.new_fpl_data.iat[i, 7] += 1
                            elif self.new_fpl_data.iat[i, 2] == u: #midのときはmid, app, arrを遅延指示
                                self.new_fpl_data.iat[i, 5] += 1
                                self.new_fpl_data.iat[i, 6] += 1
                                self.new_fpl_data.iat[i, 7] += 1
                            elif self.new_fpl_data.iat[i, 3] == u: #appのときはapp, arrを遅延指示
                                self.new_fpl_data.iat[i, 6] += 1
                                self.new_fpl_data.iat[i, 7] += 1
                            

                            self.total_delay[u] += 1
                            if np.all(self.num_delay[:,u] != i): 
                                self.uatm_data = np.append(self.uatm_data, self.aircraft_num.reshape(1, 4), axis=0)
                                self.num_delay = np.append(self.num_delay, axis=0)
                            if self.new_flight_data[t, i] != 10: self.total_air_delay[u] += 1 #空中遅延なら＋１
                    else:
                        pass
                self.aircraft_num = np.array([np.count_nonzero(self.new_flight_data[t] == 0), 
                                          np.count_nonzero(self.new_flight_data[t] == 1), 
                                          np.count_nonzero(self.new_flight_data[t] == 2), 
                                          np.count_nonzero(self.new_flight_data[t] == 3)])
        return self.new_fpl_data, self.new_flight_data

        
        
class FlightPlan:
    def __init__(self, uatm_list, max_time, max_flight_time):
        self.num_uatm = len(uatm_list)
        self.uatm_list = uatm_list
        self.max_time = max_time
        self.max_flight_time = max_flight_time
        #self.flight_num = flight_num
    
    def making_fpl(self, i):
        self.dep, self.arr = random.sample(self.uatm_list, 2)
        self.uatm_other = copy.copy(self.uatm_list)
        self.uatm_other.remove(self.dep)
        self.uatm_other.remove(self.arr)

        self.dep_time = random.randint(0, self.max_time-self.max_flight_time)
        if self.dep + self.arr == 3:
            self.mid = random.choice(self.uatm_other)

            self.arr_time = random.randint(self.dep_time+900, self.dep_time+self.max_flight_time)
            self.mid_time = random.randint(self.dep_time+300, self.arr_time-600)
            self.app_time = random.randint(self.mid_time+300, self.arr_time-300)
            self.data = pd.DataFrame([[i, self.dep, self.mid, self.arr, self.dep_time, self.mid_time, self.app_time, self.arr_time]], 
                         columns=["flight_id", "dep", "mid", "arr", "dep_time", "mid_time", "app_time", "arr_time"])
            return self.data
           
        else:
            self.arr_time = random.randint(self.dep_time+600, self.dep_time+self.max_flight_time)
            self.app_time = random.randint(self.dep_time+300, self.arr_time-300)
            self.data = pd.DataFrame([[i, self.dep, self.arr, self.dep_time, self.app_time, self.arr_time]], 
                         columns=["flight_id", "dep", "arr", "dep_time", "app_time", "arr_time"])
            return self.data
        
        
        
        
        
class FlightWorld:
    def __init__(self, fpl_data, max_time):
        self.fpl_data = fpl_data
        self.flight_num = len(self.fpl_data)
        self.max_time = max_time
        
        self.current_aircraft_num = np.zeros(4)
        self.current_aircraft_callsign = [[]]
        
    def flight_step(self, time, uatm_onetime_data):
        self.time = time 
        self.uatm_onetime_data = uatm_onetime_data
        for i in range(self.flight_num):
            if self.fpl_data.iat[i, 4] == self.time:                           #t=DEPtimeのとき
                self.uatm_onetime_data[i] = self.fpl_data.iat[i, 1]
            elif self.fpl_data.iat[i, 5] == self.time:                          #t=MIDtimeのとき
                self.uatm_onetime_data[i] = self.fpl_data.iat[i, 2]
            elif self.fpl_data.iat[i, 6] == self.time:                          #t=APPtimeのとき
                self.uatm_onetime_data[i] = self.fpl_data.iat[i, 3]
            elif self.fpl_data.iat[i, 7] == self.time:                          #t=ARRtimeのとき
                self.uatm_onetime_data[i] = 100
            else:
                pass
        return self.uatm_onetime_data
    
    
    def flight(self, flight_num, sim_time):
        self.flight_data = np.zeros((0, flight_num)) #number of flight

        self.uatm_onetime_data = np.full(flight_num, 10)
        self.uatm_data = np.zeros((0, 4))

        for time in range(sim_time): 
            self.uatm_onetime_data = self.flight_step(time, self.uatm_onetime_data)
            self.flight_data = np.append(self.flight_data, self.uatm_onetime_data.reshape(1, flight_num), axis=0)
    
            self.aircraft_num = np.zeros(4)
            for n in range(flight_num):
                if self.uatm_onetime_data[n] == 0:
                    self.aircraft_num[0] += 1
                elif self.uatm_onetime_data[n] == 1:
                    self.aircraft_num[1] += 1
                elif self.uatm_onetime_data[n] == 2:
                    self.aircraft_num[2] += 1
                elif self.uatm_onetime_data[n] == 3:
                    self.aircraft_num[3] += 1
            self.uatm_data = np.append(self.uatm_data, self.aircraft_num.reshape(1, 4), axis=0)
            
        return self.flight_data, self.uatm_data
```

```python
#fpl作成
flight_num = 100
sim_time = 10800
max_flight_time = 3600
fpl_data = pd.DataFrame()
f = FlightPlan([0,1,2,3], sim_time, max_flight_time)
for i in range(flight_num): 
    one_fpl = f.making_fpl(i)
    fpl_data = pd.concat([fpl_data, one_fpl])
    
fpl_data.reset_index(drop=True)

original_csv = "data/original" + f"_{seed}" + f"_result_{datetime.now():%Y%m%d%H%M%S}.csv"
fpl_data.to_csv(original_csv)
fpl_data
```

```python
%%time

fw = FlightWorld(fpl_data, sim_time)
flight_data, uatm_data = fw.flight(flight_num, sim_time)
print(flight_data)
print(uatm_data)
```

```python
%%time
#DCB実行　

new_fpl_data = fpl_data.copy()
new_flight_data = flight_data.copy()

ruatm = RUATM(np.array([10, 10, 10, 10]))    #assign the capacity of each RUATM
dcb = DCB(np.array([10, 10, 10, 10]))

new_fpl_data, new_flight_data = dcb.nonDCB(new_fpl_data, new_flight_data)
```

```python
%%time

new_fw = FlightWorld(new_fpl_data, sim_time)
new_flight_data, new_uatm_data = new_fw.flight(flight_num, sim_time)
```

```python
#fig.clear()
def graph(data_name, title):
    data = data_name
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111)
    x = np.arange(10800)

    #ax.set_aspect('equal')      

    ax.set_xlim(0, 10800)                  
    ax.set_ylim(0, 15) 
    
    ax.set_xlabel("time [s]",fontsize=23)                 
    ax.set_ylabel("number of aircraft",fontsize=23)

    ax.set_xticks(np.linspace(0, 10000, 5))
    ax.set_yticks(np.linspace(0, 15, 4))
    plt.tick_params(labelsize=20)

    plt.title(title)
    plt.grid(True)
    plt.rcParams['font.family'] = 'Times New Roman' # Fonts
    plt.plot(x, data[:,0], marker="o", markersize=1, color='tab:blue', label='RUATM A')
    plt.plot(x, data[:,1], marker="o", markersize=1, color='tab:orange', label='RUATM B')
    plt.plot(x, data[:,2], marker="o", markersize=1, color='tab:green', label='RUATM C')
    plt.plot(x, data[:,3], marker="o", markersize=1, color='tab:red', label='RUATM D')

    ax.legend(bbox_to_anchor=(1, 1), loc='upper right', fontsize=15)

    flight_png = "data/" + title + f"_{seed}" + f"_result_{datetime.now():%Y%m%d%H%M%S}.png"
    plt.savefig(flight_png)
    plt.show()

graph(uatm_data, 'original')
graph(new_uatm_data, 'nonDCB')

print(dcb.total_delay) # 総遅延時間
print(dcb.total_air_delay)
```
